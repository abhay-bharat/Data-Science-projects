# Training the model

```python
# Make a list of all columns that have categorical features to apply 1-hot encoding
count = 0
columns = []
for col in train_data.columns:
    if type(train_data[col][0]) == str:
        count = count + 1
        columns.append(col)
        print(col)
print(count)
```

```python
# Write function to perform one hot encoding
def category_onehot(multiCols):
    df_final = final_df
    i = 0
    for fields in multiCols:
        print(fields)
        df1 = pd.get_dummies(final_df[fields], drop_first= True)
        
        final_df.drop([fields], axis = 1, inplace = True)
        if i == 0:
            df_final= df1.copy()
        else:
            df_final = pd.concat([df_final, df1], axis = 1)
        i = i+1
        
    df_final = pd.concat([final_df, df_final], axis = 1)
    return df_final
```

```python
# lets keep copy of train data
train = train_data.copy()
```

```python
# before encoding we will have to concat train and test because, there are some 
# category that don't appear in train but do in test
final_df = pd.concat([train_data, test_data], axis = 0)
```

```python
final_df.shape
```

```python
# now lets apply 1-hot encoding
final_df = category_onehot(columns)
```
