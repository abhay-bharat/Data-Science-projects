```python
# Car name is not the feature required, as mathematical model has selling_price
df_final = data[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
df_final
```

## Creating a  new derived feature, number of years

```python
# Creating a feature to know how old the car is
df_final['Current_year'] = 2020
```

```python
df_final.head(5)
```

```python
# Subtracting 2020 - Year
df_final['num_year'] = df_final['Current_year'] - df_final['Year']
```

```python
df_final.head(5)
```

```python
#### Now current year and year feature is not required 
df_final.drop(['Year', 'Current_year'], axis = 1, inplace = True)
```

```python
# Now perform one-hot encoding 
df_final = pd.get_dummies(df_final, drop_first = True)
```

```python
df_final.head()
```

```python
# Selling price is dependent faeture which is to be predicted
# X is the features from which we have to predict 
X = df_final.iloc[:,1:]
y = df_final.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

# To know importance of each feature 

```python
# ExtraTreeRegressor uses ensemble technique, (related to random forest regressor)
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
i = 0
for imp in model.feature_importances_:
    print(X.columns[i],"has feature importance of :", imp)
    i +=1
```
