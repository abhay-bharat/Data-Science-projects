## Questions to explore:
- **How does correlation between feartures affect model and why it is not desired to have a high correlation between features?**
Ans: 'Multicolinearity', The prediction of the model (often denoted as y-hat) can still be accurate when multicollinearity is present among the predictor variables. However, the estimate of the coefficients for those correlated predictors is not likely to be meaningful.It will make the estimate highly in-stable. This instability will increase the variance of estimates. It means that if there is a small change in X, produces large changes in estimate.

- **How skewness in features affect the model?**
Ans : One reason could be that skewness could lead to biased predictions. That is high bias 



## visualising data, Univariate plots(visualising the trend of a single variable)

```python
#plotting histogram 
df.hist()
```

```python
### Density plot, better visual substitute for histogram as considers density distribution and not in bins frequency
### When dealing with subplots it is best to have sharex to be False, which means all plots do not share same x-scale values.
df.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
```

```python
### Plotting box plot to visualize the distribution around central tendencies. Also to look at the potential outliers
df.plot(kind= 'box', subplots = True, layout = (3,3),sharex = False, sharey = False)
```

## Multivariate plots, visualising the trend between the multiple variable in a dataset

```python
### Correlation matrix represented using a heat map
correlations = df.corr()
#plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin = -1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0,9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
names = df.columns
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
